```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import os
import sys

# Add the parent directory to sys.path (to import util function)
parent_dir = os.path.abspath(os.path.join(os.getcwd(), ".."))
if parent_dir not in sys.path:
    sys.path.append(parent_dir)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


df_all = pd.read_csv("../data_host/all_gene_annotations.csv",sep=",",low_memory=False)
df_all['biome'] = df_all['biome'].fillna(-1)
df_all['mammal_or_not'] = df_all['mammal_or_not'].fillna(-1)
df_all['host_or_not'] = df_all['host_or_not'].fillna(-1)

df_train = pd.read_csv("../data_host/all_gene_annotatations.training.tsv",sep="\t",low_memory=False)
df_train['biome'] = df_train['biome'].fillna(-1)
df_train['mammal_or_not'] = df_train['mammal_or_not'].fillna(-1)
df_train['host_or_not'] = df_train['host_or_not'].fillna(-1)

df_test = pd.read_csv("../data_host/all_gene_annotations.testing.tsv",sep="\t",low_memory=False)
df_test['biome'] = df_test['biome'].fillna(-1)
df_test['mammal_or_not'] = df_test['mammal_or_not'].fillna(-1)
df_test['host_or_not'] = df_test['host_or_not'].fillna(-1)
```

```python
from collections import defaultdict

def create_dict(unique_labels):
    diction = defaultdict()
    for i, label in enumerate(unique_labels):
        diction[label] = i
    diction[-1] = -1    
    return diction    

def count_dict(df, column_name):
    counts = defaultdict(int)
    for i in range(len(df[column_name])):
        counts[df_all[column_name][i]] += 1
    return counts    

host_or_not_uniq = np.unique([bi for bi in df_all['host_or_not'] if isinstance(bi, str)])
host_or_not_dict = create_dict(host_or_not_uniq)

mammal_or_not_uniq = np.unique([bi for bi in df_all['mammal_or_not'] if isinstance(bi, str)])
mammal_or_not_dict = create_dict(mammal_or_not_uniq)

biome_uniq = np.unique([bi for bi in df_all['biome'] if isinstance(bi, str)])
biome_dict = create_dict(biome_uniq)

host_or_not = [1 if ci == 'ecological' else 0 for ci in df_all['host_or_not']]
mammal_or_not = [1 if ci == 'mammal' else 1 for ci in df_all['mammal_or_not']]


host_or_not_counts_train = count_dict(df_train, 'host_or_not')
host_or_not_counts_test = count_dict(df_test, 'host_or_not')

mammal_or_not_counts_train = count_dict(df_train, 'mammal_or_not')
mammal_or_not_counts_test = count_dict(df_test, 'mammal_or_not')

biome_counts_train = count_dict(df_train, 'biome')
biome_counts_test = count_dict(df_test, 'biome')


plt.figure(figsize=(12.5, 3))
val = list(biome_counts_train.values())
key = list(biome_counts_train.keys())
plt.bar(range(len(val[1:])), val[1:])
plt.xticks(range(len(val[1:])), key[1:], rotation=90);
```
